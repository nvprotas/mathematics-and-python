```python
from math import sin
from math import exp

import numpy
import matplotlib as mpl
import matplotlib.pyplot as plt
% matplotlib inline
mpl.rc('font', family='Verdana', size= 16)

def func(x):
    return  sin(x / 5.) * exp(x / 10.) + 5 * exp(-x / 2.)

fig, ax = plt.subplots(figsize=(10,5))
x = numpy.linspace(0,20,200)

y1 = map(lambda x: func(x), x)
ax.plot(x, y1, label = 'func')


A2 = numpy.array([[1., 1.], [1., 15.]])
B2 = numpy.array([func(1),func(15)])
Sol2 = numpy.linalg.solve(A2, B2)

y2 = map(lambda x: Sol2[0] + Sol2[1]*x, x)
ax.plot(x, y2, label = 'pow2')

A4 = numpy.array([[1., 1., 1. ,1.],
                  [1., 4., 4**2, 4**3],
                  [1., 10., 10**2, 10**3],
                  [1., 15., 15**2, 15**3]])
B4 = numpy.array([func(1),func(4),func(10),func(15)])
Sol4 = numpy.linalg.solve(A4, B4)
y4 = map(lambda x: Sol4[0] + Sol4[1]*x + Sol4[2]*x*x + Sol4[3]*x*x*x, x)
ax.plot(x, y4, label = 'pow4')

ax.legend(bbox_to_anchor=(1., 1.))

print Sol4.round(2)
```
