```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from math import sqrt
%matplotlib inline
```

# Распределение Лапласа

Плотность вероятности: $f(x)\;=\;\frac1{2b}\exp\left(\mathit-\frac{\left|\mathit x\mathit\;\mathit-\mathit\;\mathit\mu\right|}{\mathit b}\right) $

Мат. ожидание:$\mu$, дисперсия: $2b^2 $

Сгенерируем выборку с параметрами $\mu = 2$, $b = 0.5$

```python
loc = 2
scale = 0.5
laplace = sts.laplace(loc,scale)
sample = laplace.rvs(1000)
```

Построим гистограмму выборки и график теоретической плотности распределения

```python
x = np.linspace(loc-3,loc+3,100)
pdf = laplace.pdf(x)
plt.plot(x, pdf)
plt.hist(sample,bins=15, normed=True)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Далеее сгенерированы 1000 выборок при различных объемах выборок n = (3, 5, 10, 50) и построены гистограммы распределений их выборочных средних. Поверх гистограмм построены плотности соответствующих нормального распределения c параметрами $N(\mu, \frac{2b^2}{n} ) $

```python
n = 3
vyborka =  []
for k in range(1000):
    smpl = laplace.rvs(n)
    vyborka.append(smpl.mean())

norm_rv = sts.norm(loc,sqrt(2*scale*scale/n))
x = np.linspace(loc-3,loc+3,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf)
plt.hist(vyborka,bins=20,normed=True)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.draw()
```
